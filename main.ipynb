```python
print(df['mainroad'].unique())
print(df['guestroom'].unique())
print(df['basement'].unique())
print(df['hotwaterheating'].unique())
print(df['airconditioning'].unique())
print(df['prefarea'].unique())
print(df['furnishingstatus'].unique())
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
```

```python
mainroad = np.array(df['mainroad']).reshape(-1,1)
guestroom = np.array(df['guestroom']).reshape(-1,1)
basement = np.array(df['basement']).reshape(-1,1)
hotwaterheating = np.array(df['hotwaterheating']).reshape(-1,1)
airconditioning = np.array(df['airconditioning']).reshape(-1,1)
prefarea = np.array(df['prefarea']).reshape(-1,1)


df['mainroad']=encoder.fit_transform(mainroad)
df['guestroom']=encoder.fit_transform(guestroom)
df['basement']=encoder.fit_transform(basement)
df['hotwaterheating']=encoder.fit_transform(hotwaterheating)
df['airconditioning']=encoder.fit_transform(airconditioning)
df['prefarea']=encoder.fit_transform(prefarea)
```

```python
furnishingstatus = np.array(df['furnishingstatus']).reshape(-1,1)

df['furnishingstatus']=encoder.fit_transform(furnishingstatus)
```

Splitting the Dataset

```python
features=['area','mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.loc[:, df.columns != 'price'], df['price'], test_size=0.2, random_state=42)
```

```python
X_train.info()
```

```python
y_train.info()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()   

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_train_log = np.log1p(y_train)  
y_test_log = np.log1p(y_test)
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
```

```python
rf = RandomForestRegressor(n_estimators=150)
rf.fit(X_train_scaled, y_train_log)
```

```python
from sklearn.metrics import mean_squared_error, adjusted_rand_score, r2_score 
```

```python
y_pred= rf.predict(X_test_scaled)
```

```python
print(r2_score(y_test_log, y_pred))
print(mean_squared_error(y_test_log, y_pred))
```

Feature Importance

```python
plt.bar(range(X_train.shape[1]), rf.feature_importances_)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()
```
